```python
import glob 
import pandas as pd
```

```python
# Change this to the directory where your files are stored.
# Here, I'm only looking at puzzle1
filenames = glob.glob("/Users/Chloe/OneDrive - University of Toronto/talk2me_jr/puzzle_solving/puzzle1/*.txt")
print(filenames[:10])
```

```python
puzzle_results = []
for filename in filenames: # Loop through every filename
    with open(filename) as f: # Open the file
        response = f.read() # Read in the response
    # Update your puzzle_results list
    puzzle_results += [{'filename': filename, 'response': response}]
    
# We saved our responses as a list of objects
print(puzzle_results[:10])
```

```python
# The list of objects is easy to convert to a pandas DataFrame
df = pd.DataFrame(puzzle_results)
df
```

```python
# Let's modify our DataFrame a bit
# From the filename, we extract the user_id and session_id
# Split by "/" and retain the last element
# Then split by "_" and retain the elements which correspond to the user_id and session_id
df['user_id'] = [filename.split('/')[-1].split('_')[0] for filename in df['filename']]
df['session_id'] = [filename.split('/')[-1].split('_')[1] for filename in df['filename']]
df
```

```python
df['puzzle'] = 1 # This is for puzzle 1
df
```
